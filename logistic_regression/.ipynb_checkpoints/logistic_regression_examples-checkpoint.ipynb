```python
test = pd.read_csv('titanic_test.csv')
```

```python
def imputAge(cols):
    Age = cols[0]
    PClass = cols[1]
    if pd.isnull(Age):
        if PClass ==1:
            return 37
        elif PClass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass']].apply(imputAge,axis=1)
```

```python
sns.heatmap(train.isnull(),yticklabels = False, cbar=False,cmap='viridis')
```

```python
train.drop('Cabin',axis =1 , inplace=True)
```

```python
sns.heatmap(train.isnull(),yticklabels = False, cbar=False,cmap='viridis')
```

```python
train.dropna(inplace=True)
```

```python
sns.heatmap(train.isnull(),yticklabels = False, cbar=False,cmap='viridis')
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
```

```python
embarked = pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
train2=pd.concat([train,sex, embarked], axis = 1)
```

```python
train2.head()
```

```python
train2.columns
```

```python
train2.drop(['Sex','Name','Embarked','Ticket'],inplace=True,axis=1)
```

```python
train2.drop(['PassengerId'],inplace=True,axis=1)
```

```python
Pclass = pd.get_dummies(train2['Pclass'],drop_first=True)
```

```python
train3 = pd.concat([train2,Pclass],axis=1)
```

```python
train3.drop(['Pclass'],inplace=True,axis=1)
```

```python
train3.head()
```

```python
X = train3
```

```python
X
```

```python
Y = train2['Survived']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,Y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(Y_test,predictions))
```
