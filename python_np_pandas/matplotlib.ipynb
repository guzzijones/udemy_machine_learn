```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x **2
```

```python
x
```

```python
y
```

```python
#functional
plt.plot(x,y,'r-')
plt.xlabel("xvalues")
plt.ylabel("Yvalues")
plt.title("title")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b-')
```

```python
#object oriented method
#create figure objects
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([.1,.5,.2,.5])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title("small")
axes1.set_title("large")
```

```python
axes1.plot(x,y)
```

```python
fix, axes =  plt.subplots(nrows=1,ncols=2)
for cur in axes:
    cur.plot(x,y)
```

```python
fix, axes =  plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("Title")
plt.tight_layout()
```

```python
fig =plt.figure(figsize=(3,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols =1,figsize=(3,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('mypic.pdf')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.set_title("test")
ax.set_ylabel("ylabel")
ax.set_xlabel("xLabel")
#ax.legend(loc=10)
ax.legend(loc=(.1,.1))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=5,linestyle='--',marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
ax.set_xlim(0,1)
ax.set_ylim(0,2)
```
