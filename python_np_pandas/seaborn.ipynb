```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# dist plot - univariate observations

```python
sns.distplot(tips['total_bill'],kde=False)
```

# join plot  - bivariate data

```python
sns.jointplot(x ='total_bill' ,y ='tip' , data=tips)
```

```python
#two histograms and a scatter. as bill gets higher bill gets higher
sns.jointplot(x ='total_bill' ,y ='tip' , data=tips,kind='hex')
```

```python
sns.jointplot(x ='total_bill' ,y ='tip' , data=tips,kind='reg')
```

```python
sns.jointplot(x ='total_bill' ,y ='tip' , data=tips,kind='kde')
```

```python
#entire data frame # use hue to color categorical.
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

# regression plots

```python

sns.lmplot(x='total_bill',y='tip',data=tips,scatter_kws={'s':100},hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',aspect=0.6,size=8)
```
